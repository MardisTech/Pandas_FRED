```python
#looking thru fred for unemployment rates by state updated monthly
fred.search('unemployment rate state', filter=('frequency', 'Monthly'))
```

```python
#saving to an object
unemp_df = fred.search('unemployment rate state', filter=('frequency', 'Monthly'))
```

```python
#checking first few rows
unemp_df.head()
```

```python
#filter for seasonal adjustment and units
unemp_df = unemp_df.query('seasonal_adjustment == "Seasonally Adjusted" and units == "Percent"')
```

```python
#continue filtering title column for 'unemployment rate'
unemp_df = unemp_df.loc[unemp_df['title'].str.contains('Unemployment Rate in')]
```

```python
unemp_df.head()
```

```python

all_results = []

#looping through the index's of unemp_df to add the coresponding series data to all_results array. will be an array of data frames
for myid in unemp_df.index:
    results = fred.get_series(myid)
    results = results.to_frame(name=myid) #by default data fram created with name=0
    all_results.append(results)
    #list of data frames
```

```python
all_results[2]
```

```python
#concat will accumulate all dataframes in list into a single data frame, then dropping two columns that had Nan data so we can filter out Nan in next step. if we didnt remove these columns then all rows would have a Nan value and be removed from next step
uemp_results = pd.concat(all_results, axis=1)
```

```python
#drop all rows with Nan anywhere as value
uemp_states = uemp_results.dropna()
```

```python
#plot as line plot, name of uemp_states
px.line(uemp_states)
```

```python
# Noticing the large spike in 2020 we can investigate this time period further
```

```python
#the df is indexed by date, so we can locate (loc) where the index is equal to march 2020
#suspiciously, the UNEM data for march and april of 2020 is missing
# .T transposes the rows to columns then we sort lowest to highest
uemp_states.loc[uemp_states.index == '2020-05-01'].T. \
    sort_values('2020-05-01') \
    .plot(kind='bar', figsize=(20, 5), title="Unemployment Rate by Stat, May 2020 (percent)")
```

```python
#we can see that the few columns may not be state data judging by the size. we also noticethe state data names on the x axis are not very intuitive. We will attempt to convert the state id's to readable values
```

```python

#we can see the title has "unemployment rate in [state name]"
unemp_df['title'].str.replace('Unemployment Rate in', '').to_dict()
# we can remove that phrase from title leaving only the state name, then convert to dictionary to map id's to state names
id_to_state = unemp_df['title'].str.replace('Unemployment Rate in', '').to_dict()
id_to_state
```

```python
uemp_states.columns
```

```python
#update the uemp_states.columns to a list that gives the key to the state id from dictionary
uemp_states.columns = [id_to_state[val] for val in uemp_states.columns]
```

```python
px.line(uemp_states)
```

```python
ax = uemp_states.loc[uemp_states.index == '2020-05-01'].T. \
    sort_values('2020-05-01') \
    .plot(kind='barh', figsize=(8, 12), width=0.9, title="Unemployment Rate by State, May 2020 (percent)")
ax.set_xlabel('% Unemployed')
plt.show()
```
